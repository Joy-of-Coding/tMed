```python
# img_path = os.path.join('data', 'images', 'bike.png')
img_path = os.path.join('data', 'images', 'CanField', 'ISIC_0000000.jpg')
```

```python
from matplotlib import pyplot as plt
```

```python
# Show image
plt.imshow(img)
```

```python
# reOrder Color Channels from OpenComputerVision to MatPlot Library
recolor = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```
